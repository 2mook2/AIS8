```python
raw.info()
```

```python
last_timestamp = raw['event_time'].max() + dt.timedelta(days=1)
last_timestamp
```

```python
rfm = raw.groupby('user_id').agg(
    Recency=('event_time', lambda x : (last_timestamp - x.max()).days),
    Frequency=('user_id', 'count'), # 구매 건수지, 구매 물품수가 아니다. 구매 물품수라면 Frequency=('user_session', 'sum')
    Monetary=('price', 'sum')    
)

rfm
```

```python
rfm.hist(bins=50)
```

```python
rfm.describe()
```

```python
rfm.quantile(q=[0.25,0.5,0.75])
```

```python
quartiles=rfm.quantile(q=[0.25,0.5,0.75]).to_dict()
quartiles
```

```python
## for Recency 

def R(x, q, r):
    if x <= q[r][0.25]:
        return 4
    elif x <= q[r][0.50]:
        return 3
    elif x <= q[r][0.75]: 
        return 2
    else:
        return 1
    
## for Frequency and Monetary 

def FM(x,q,fm):
    if x <= q[fm][0.25]:
        return 1
    elif x <= q[fm][0.50]:
        return 2
    elif x <= q[fm][0.75]: 
        return 3
    else:
        return 4    
```

```python
rfm['Recency'].apply(R, args=(quartiles, 'Recency'))
rfm['Frequency'].apply(FM, args=(quartiles, 'Frequency'))
```

```python
#create RFM segmentation column
rfm['R_Quartile'] = rfm['Recency'].apply(R, args=(quartiles, 'Recency'))
rfm['F_Quartile'] = rfm['Frequency'].apply(FM, args=(quartiles, 'Frequency'))
rfm['M_Quartile'] = rfm['Monetary'].apply(FM, args=(quartiles, 'Monetary'))
rfm['RFM_segmentation'] = rfm.R_Quartile.map(str) \
                    + rfm.F_Quartile.map(str) \
                    + rfm.M_Quartile.map(str)
rfm['RFM_score'] = rfm.R_Quartile.map(int) + rfm.F_Quartile.map(int) + rfm.M_Quartile.map(int)
rfm.head()
```
