### 이항 분포

```python
from scipy.stats import binom # binom 이항 분포
```

```python
data = binom.rvs(n=100, p=0.3, size=365) # rvs = Random variates = 확률변수
# 100명의 고객, 30% 구매율, 365일 시뮬레이션
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.hist(data, bins=40, range=[10, 50]) 
```

```python
# 확률질량함수 pmf
binom.pmf(k=30, n=100, p=0.3) # 구매율이 30%, 100명 방문 고정, 그 중에 딱 30명이 구매할 확률

# => 그런데 고객 개개인의 구매확률은 30% 지만 전체로는 딱 30% 이 아님을 알 수 있다.
```

```python
# 누적분포함수
binom.cdf(k=30, n=100, p=0.3) # 0 ~ 30명까지 구매할 확률의 누적

# pmf의 결과를 보면 k는 고정되지 않고 매일매일 변동함을 알 수 있다.
```

```python
0.5491236007687914 * 365 # 직원 한 명이 감당할 수 있는 고객이 10명일 때 직원이 3명이라면 200일 정도 밖에 감당 못 함.
```

```python
binom.cdf(k=50, n=100, p=0.3) * 365 # 직원이 5명은 있어야 됨을 알 수 있다.
```

```python
# pmf나 cdf에서 k값이 늘어날 수록 확률이 커지는 지 다시 한번 설명해주시면 감사하겠습니다.
import numpy as np
x = np.arange(0, 101)
p = binom.pmf(k=x, n=100, p=0.3) # 확률이 30 이므로 30까지 커지다가 30이 넘어가면 다시 낮아진다.
plt.plot(x, p)
plt.xlabel('k')
```

```python
# 누적의 경우
p = binom.cdf(k=x, n=100, p=0.3)
plt.plot(x, p)
plt.xlabel('k')
```

### 정규 분포

```python
from scipy.stats import norm # norm 정규 분포
```

```python
m = 100 * 0.3
s = np.sqrt(100 * 0.3 * 0.7) # 루트를 씌워줘야 함
data = norm.rvs(loc=m, scale=s, size=365) 
```

```python
plt.hist(data, bins=40, range=[10, 50]) # 위에서는 이항 분포 계산 아래는 정규 분포로 근사
```

```python
norm.cdf(x=30, loc=m, scale=s)
```

```python
binom.cdf(k=30, n=100, p=0.3)
```

```python
binom.cdf(k=300, n=1000, p=0.3)
```

```python
binom.cdf(k=3000, n=10000, p=0.3) # 숫자가 점점 커질 수록 정규분포와 가까워 짐을 알 수 있다.
```
